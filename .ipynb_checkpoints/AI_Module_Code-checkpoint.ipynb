### Data Exploration

First, I explored the dataset. My initial step was to load the directories and view the existing structure for the dataset.

```python
def visualize_folder_structure(path, indent=0, file_limit=10):
    items = os.listdir(path)
    for i, item in enumerate(items):
        if i >= file_limit:
            print("  " * indent + "|-- ...")
            break
        print("  " * indent + "|-- " + item)
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            visualize_folder_structure(item_path, indent + 1, file_limit)

# Example usage
visualize_folder_structure('./data/files')
```

As you can see our classes are simply separated one folder for each file. So we're going to need to split the dataset into training, testing and validation subsets so that we can effectively work with the model. Before doing that we'll need to define any transformations to the model for the training versus the testing and validation. Both sets will need to be resized (larger images means model takes longer to train.) The testing and validation will simply need to be converted to a tensor once it has been resized. Comparatively we may wish to apply data augmentation to the training set to make it more robust. 

I have chosen to implement a random rotation of the image and a random horizontal flip to the image. Let's prepare our dataset with these transforms now.

### Dataset Preparation

Firstly we need to define our transforms as dicsussed.

```python
train_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation((45)),
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
])
```

```python
test_val_transforms = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
])
```

Then we can load the dataset. The inbuilt ImageFolder functionality in pytorch will automatically interpret the data directory structure for its labels and ignore the DICOM files.

```python
dataset = ImageFolder("./data/files")
```

As is standard in machine learning we'll then split the dataset into subsets for training validation and testing.

```python
# Define train, testing, and validation splits
train_size = int(0.6 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
```

```python
# Set the seed and create a generator object with the seed for reproducibility
seed = 123
generator = torch.Generator().manual_seed(seed)
```

```python
# Split dataset
dataset_train, dataset_val, dataset_test = random_split(dataset, [train_size, val_size, test_size], generator=generator)
```

Now we'll apply those transforms and get the dictionary for our class indexes.

```python
# Apply different transforms for train and val/test datasets
dataset_train.dataset.transform = train_transforms
dataset_val.dataset.transform = test_val_transforms
dataset_test.dataset.transform = test_val_transforms
```

```python
# Invert the class_to_idx dictionary to map index to class
idx_to_class = {v: k for k, v in dataset.class_to_idx.items()}
print(idx_to_class)
```

Finally, we'll instantiate the data loader objects from our datasets.

```python
# Create dataloaders
dataloader_train = DataLoader(dataset_train, shuffle=True, batch_size=5)
dataloader_val = DataLoader(dataset_val, shuffle=True, batch_size=5)
dataloader_test = DataLoader(dataset_test, shuffle=False, batch_size=5)
```

Let's take a look at some of the examples in our training dataloader.

```python
for i, (images, labels) in enumerate(dataloader_train):
    # Limit to 3 sets of 5
    if i >= 3:
        break
    
    # Process each image in the batch individually
    for idx in range(images.size(0)):
        image = images[idx]
        label = labels[idx]
        
        # Remove batch dimension and reshape for display
        image_np = image.permute(1, 2, 0).numpy()
        # Scale back to [0, 255] range
        image_np = (image_np * 255).astype('uint8')  
        
        # Convert to PIL and display
        image_pil = Image.fromarray(image_np)
        display(image_pil)
        print(idx_to_class[label.item()])
        print("----------")
```

## Methods

### CNN Model

#### Model Setup

A CNN model can be broadly divided into a feature extractor and a classifier for this use case. We use convolutional layers to reduce the complexity of the input image while preserving essential spatial information and local features. Following this, we apply a ReLU activation function to introduce non-linearity, enabling the model to learn more complex patterns. Finally, a MaxPooling layer is used to downsample the feature maps output from the convolution, reducing their dimensionality and computational load while retaining the most important features. This is repeated a few times and then the output is flattened so that it can be fed into the fully connected linear layer for classification.

We'll include some dropout to ensure the network is robust, isn't overly reliant on specific neurons and doesn't overfit.

```python
class CNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        # Define feature extractor
        self.feature_extractor = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(32, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(16, 8, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Flatten(),
        )
        # Define classifier
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(8 * 16 * 16, num_classes)
        )
    
    def forward(self, x):
        x = self.feature_extractor(x)
        x = self.classifier(x)
        return x
```

With the architecture built out we'll instantiate the model, the loss criterion and the optimizer.

```python
# Define the model
cnn_net = CNN(num_classes=3)
# Define the loss function
criterion = nn.CrossEntropyLoss()
# Define the optimizer
optimizer = optim.Adam(cnn_net.parameters(), lr=0.001)
```

#### Model Pre-evaluation

Lets evaluate metrics for the model before training.

```python
def evaluate_model(model, dataloader, num_classes=3):
    # Initialize metrics
    metric_precision = Precision(task="multiclass", num_classes=num_classes, average="macro")
    metric_recall = Recall(task="multiclass", num_classes=num_classes, average="macro")
    conf_matrix = ConfusionMatrix(task="multiclass", num_classes=num_classes)

    # Set model to evaluation mode
    model.eval()

    with torch.no_grad():
        for images, labels in dataloader:
            outputs = model(images)
            _, preds = torch.max(outputs, 1)

            # Update metrics
            metric_precision.update(preds, labels)
            metric_recall.update(preds, labels)
            conf_matrix.update(preds, labels)

    # Compute final metrics
    precision = metric_precision.compute()
    recall = metric_recall.compute()
    conf_matrix = conf_matrix.compute()

    return precision, recall, conf_matrix
```

```python
precision, recall, conf_matrix = evaluate_model(cnn_net, dataloader_test)
```

```python
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

Model is basically just guessing. Let's now train it for 10 epochs.

#### Model Training

Below you'll see our training loop. We ensure the model is in training mode. Calculate our loss against the training set and use this to update the parameters. Then we set it to eval mode (meaning the dropout layer isn't active) then calculate its loss against the validation set. We then put it back in training mode and repeat.

```python
def train_model(model, dataloader_train, dataloader_val, criterion, optimizer, num_epochs=5):
    for epoch in range(num_epochs):
        running_loss = 0.0
        
        # Training phase
        model.train()
        for images, labels in dataloader_train:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        epoch_loss = running_loss / len(dataloader_train)
        
        # Validation phase
        model.eval()
        validation_loss = 0.0
        # Disable gradient calculations
        with torch.no_grad():  
            for images, labels in dataloader_val:
                outputs = model(images)
                loss = criterion(outputs, labels)
                validation_loss += loss.item()
        
        epoch_val_loss = validation_loss / len(dataloader_val)
        print(f"EPOCH {epoch+1} || Training Loss: {epoch_loss:.4f}, Validation Loss: {epoch_val_loss:.4f}")
```

```python
train_model(cnn_net, dataloader_train, dataloader_val, criterion, optimizer, num_epochs=10)
```

You should be able to see both the training and validation loss decrease across epochs.

#### Results

Now let's get our metrics using our trained model.

```python
precision, recall, conf_matrix = evaluate_model(cnn_net, dataloader_test)
```

```python
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

As you can see it performs really well, we'll discuss this in the results. (Remember to save the model.)

### MLP Model

#### Model Setup

The MLP model begins by flattening the input image into a one-dimensional vector, as MLPs are designed to process vectorised data and do not retain spatial relationships.

The first fully connected layer reduces the dimensionality of the input, transforming it into a smaller and more computationally efficient size. A ReLU activation function is applied to introduce non-linearity, enabling the model to identify and learn complex patterns within the data. The second fully connected layer then maps these extracted features to the predefined output classes otherwise known as classification.

To enhance the models robustness and reduce the risk of overfitting, a dropout layer is incorporated during training for the same reasons we did in the CNN.

```python
# Define the MLP Model
class MLP(nn.Module):
    def __init__(self, num_classes):
        super(MLP, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(128*128*3, 8)
        self.fc2 = nn.Linear(8, num_classes)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.5)
    
    def forward(self, x):
        x = self.flatten(x)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.dropout(x)
        return x
```

As usual we'll instantiate our model, loss criterion and optimizer.

```python
mlp_net = MLP(3)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(mlp_net.parameters(), lr=0.001)
```

#### Model Pre-evalutation

We'll again evaluate our metrics prior to testing to ensure no funny business.

```python
precision, recall, conf_matrix = evaluate_model(mlp_net, dataloader_test)
```

```python
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

## Discussion

**So why does a CNN model generally outperform an MLP model?**

"The distinction of CNN from a plain multilayer perceptron (MLP) network is its usage of convolutional layers, pooling, and non-linearities such as tanh, sigmoid, and ReLU." Agarap, A.F.M. (2017)

The quote above underscores a key limitation of MLPs: their fully connected layers are too simplistic to process complex patterns effectively. In contrast, CNNs use convolutional filters that slide over input images, enabling downsampling while preserving and capturing essential features. As this filter (sometimes known as a window) is limited to specific areas of the image it identifies "local" features specifically.     
This approach allows CNNs to excel in tasks like image classification, as they understand spatial hierarchies and these local patterns more efficiently. Additionally, their downsampling reduces computational complexity, making them more scalable for larger datasets and deeper architectures.

**How would I extend this report?**

In our comparsion of results and discussion, we determined that the size of the training dataset was small. Truthfully, the size of the entire dataset was pretty small. If I were going to extend on this report, I would extend on my implementation of data augmentation. In this report, I used the augmentation to add variety to the dataset to avoid overfitting, however, I kept the dataset the same size. What I could do is to apply all the transformations individually and include multiple variations of the same image in my dataset thus artificially increasing its size. This is called artificial data generation.    
This helps to prevent against some of the more common issues in a limited dataset: "models trained with limited data suffer from over-fitting with significantly degenerative performance on testing datasets, and models trained with unbalanced samples will lead to poor generalization ability." Zaitian Wang et al. (2024)

I'd also like to do some hyperparameter tuning. I used a pretty standard learning rate of 0.001 but there may be a more optimal value.    
Additionally, I could do some sensitivity testing on the performance of my models dependent on the training, validation and testing split. I found through trial and error that having a larger testing split was required for more accurate metrics due to the dataset being small but it would have been good to programmatically confirm this.
